```python
#select pcp file
pcp_file = input('Enter pcp file name location. Do not include the file extension. ex: ~/Downloads/qualys-pcp-hostlist:\n')
pcp_file = pcp_file + ".csv"
file_check(pcp_file,'pcp')
```

```python
#select cloud file
cloud_file = input('Enter cloud file name location. Do not include the file extension. ex: ~/Downloads/qualys-cloud-hostlist:\n')
cloud_file = cloud_file + ".csv"
file_check(cloud_file,'cloud')
```

```python
start = time.perf_counter()
#converting file to pd
cloud = pd.read_csv(cloud_file)
cloud['From']=''
for x in cloud["From"]:
    cloud["From"] = "Cloud"
pcp = pd.read_csv(pcp_file)
pcp['From']=''
for x in pcp["From"]:
    pcp["From"] = "PCP"
```

```python
print("Merging data")
merg = cloud.append(pcp)
print("Data count after merge: " + str(len(merg)))
init_count = len(merg)
time.sleep(2)
```

```python
#Adding columns
merg["IP in CMDB"]=""
merg['DNS in CMDB']=''
```

```python
#query starts
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(ip_search, merg['IP'])
    executor.map(dns_search, merg['DNS'])
```

```python
with pd.ExcelWriter('CMDB_Queried_Data_Latest_Test.xlsx') as writer:
    merg.to_excel(writer, index=False)

print("CMDB_Queried_Data_Latest_Test.xlsx has been exported")
stop = time.perf_counter()

print("It took "+str(round((stop-start)/60,2))+" minutes for the script to complete")
```
