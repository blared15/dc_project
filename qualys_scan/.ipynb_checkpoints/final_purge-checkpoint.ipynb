```python
import os, time, shutil
import pandas as pd
columns = shutil.get_terminal_size().columns
```

```python
def file_check(file_name):
    check = os.path.isfile(file_name)
    file_name = file_name +".xlsx"
    while check == False:
        file_name = input("File name not found.\n Enter filename without the file extension. ex: ~/Downloads/latest_sample_pinged_data:  ")
        check = os.path.isfile(file_name)
    print(" file found: " + str(check))    

#extracts the filtered data to its own variable
def sheet(source,column, string):
    x = source[source[column].str.contains(string,na=False)]
    return x
#removes the filtered data from the main variable
def sheet_x(source,column, string):
    x = source[~source[column].str.contains(string,na=False)]
    return x

#splits to cmdb and out in cmdb
def cmdb(x):
        ins = x[(x['IP in CMDB']=='IP in CMDB') | (x['DNS in CMDB'] == 'DNS in CMDB')]
        out = x[(x['IP in CMDB'] =='IP NOT in CMDB') & (x['DNS in CMDB'] == 'DNS NOT in CMDB')]
        return ins, out
#function for Worksheet export    
def workbooks(source, title):
    if len(source) < 1:
        print(f"{title} is empty. It will not be added to the workbook")
    else:
        source.to_excel(writer,sheet_name=title, index=False)
```

```python
title=r'''
  _____                        _           _____ _               _       
 |  __ \                      | |         / ____| |             | |      
 | |__) |   _ _ __ __ _  ___  | |_ ___   | (___ | |__   ___  ___| |_ ___ 
 |  ___/ | | | '__/ _` |/ _ \ | __/ _ \   \___ \| '_ \ / _ \/ _ \ __/ __|
 | |   | |_| | | | (_| |  __/ | || (_) |  ____) | | | |  __/  __/ |_\__ \
 |_|    \__,_|_|  \__, |\___|  \__\___/  |_____/|_| |_|\___|\___|\__|___/
                   __/ |                                                 
                  |___/  
'''
print(title)
print(''.center(columns,'*'))
print('This Script will purge workbook data to subsheets ex: Main, Agent,VPN, PDU... '.center(columns))
print('It will only take XLSX file and ideally would need to have all CMDB queries done'.center(columns))
print('For questions with the filters of the scripts you can contact Raenielle at rsunga@netsuite.com'.center(columns))
print(''.center(columns,'*'))
```

```python
file_name = input("Please enter the filename without the file extension. ex: ~/Downloads/latest_sample_pinged_data:  ")
file_name = file_name +".xlsx"
file_check(file_name)
print('loading up workbook')
start = time.perf_counter()
workbook = pd.ExcelFile(file_name)
data = pd.read_excel(workbook)
```

```python
print('Sorting "Duplicate IP }"')

duplicate = data[data.duplicated(subset=['IP'],keep='last')]
main = data[~data.duplicated(subset=['IP'],keep='last')]
```

```python
print('Sorting "Available"')
available = sheet(main,'DNS','available')
main = sheet_x(main,'DNS','available')
#CNAME Check
available = available.append(sheet(main,'CNAME','AVAILABLE'))
main = sheet_x(main,'CNAME','AVAILABLE')
```

```python
print('Sorting "Internet Facing Assets"')
pub = sheet(main,'Tags','Internet Facing Assets')
main = sheet_x(main,'Tags','Internet Facing Assets')
```

```python
print('Sorting "Load Balancers"')
lb = sheet(main,'DNS','-mip|-vlb|-plb')
main = sheet_x(main,'DNS','-mip-mip|-vlb|-plb')
#Tags
lb = sheet(main,'Tags','loadbalancer')
main = sheet_x(main,'Tags','loadbalancer')
```

```python
print('Sorting "Agent"')
agent = sheet(main,'Tracking Method','AGENT')
main = sheet_x(main,'Tracking Method','AGENT')
```

```python
print('Sorting "links"')
link = sheet(main,'DNS','-link')
main = sheet_x(main,'DNS','-link')
```

```python
print('Sorting "None"')
blank = sheet(main,'DNS','None|none')
main = sheet_x(main,'DNS','None|none')
```

```python
print('Sorting "DevOps"')
devops = sheet(main,'Tags','BU:DevOps')
main = sheet_x(main,'Tags','BU:DevOps')
```

```python
print('Sorting "Duplicate DNS"')

duplicate = duplicate.append(main[main.duplicated(subset=['DNS'],keep='last')])
main = main[~main.duplicated(subset=['DNS'],keep='first')]
```

```python
print('Sorting "LS"')
ls = sheet(main,'DNS','-ls')
main = sheet_x(main,'DNS','-ls')
```

```python
print('Sorting "CY"')
cy = sheet(main,'DNS','cy')
main = sheet_x(main,'DNS','cy')
```

```python
print('Sorting "bo1"')
bo1 = sheet(main,'DNS','BO1|bo1|b01')
main = sheet_x(main,'DNS','BO1|bo1|b01')
#Tags
bo1 = bo1.append(sheet(main,'Tags','BO1|bo1|b01'))
main = sheet_x(main,'Tags','BO1|bo1|b01')
```

```python
print('Sorting "Bronto"')
bronto = sheet(main,'DNS','Bronto|bronto')
main = sheet_x(main,'DNS','Bronto|bronto')
#tags
bronto = bronto.append(sheet(main,'Tags','Bronto|bronto'))
main = sheet_x(main,'Tags','Bronto|bronto')
```

```python
print('Sorting "OpenAir"')
openair = sheet(main,'Tags','openair|OpenAir')
main = sheet_x(main,'Tags','openair|OpenAir')
```

```python
print('Sorting "OCI"')
oci = sheet(main,'DNS','OCI|oci')
main = sheet_x(main,'DNS','OCI|oci')
#tags
oci = oci.append(sheet(main,'Tags','OCI|oci'))
main = sheet_x(main,'Tags','OCI|oci')
```

```python
print('Sorting "Payroll"')
payroll = sheet(main,'Tags','payroll|Payroll')
main = sheet_x(main,'Tags','payroll|Payroll')
```

```python
print('Sorting "VPN"')
vpn = sheet(main,'DNS','vpn|ovpn')
main = sheet_x(main,'DNS','vpn|ovpn')
#tags
vpn = vpn.append(sheet(main,'Tags','vpn|ovpn'))
main = sheet_x(main,'Tags','vpn|ovpn')
```

```python
print('Sorting "Console"')
console = sheet(main,'DNS','console')
main = sheet_x(main,'DNS','console')
#tags
console = console.append(sheet(main,'Tags','console'))
main = sheet_x(main,'Tags','console')
```

```python
print('Sorting "Interface"')
interface = sheet(main,'DNS','interface')
main = sheet_x(main,'DNS','interface')
```

```python
print('Sorting "NoUse"')
nouse = sheet(main,'DNS','nouse')
main = sheet_x(main,'DNS','nouse')
```

```python
print('Sorting "VM"')
vm = sheet(main,'DNS','snap|\.f\.|\.m\.')
main = sheet_x(main,'DNS','snap|\.f\.|\.m\.')
```

```python
print('Sorting "PDU"')
pdu = sheet(main,'DNS','pdu-')
main = sheet_x(main,'DNS','pdu-')
#IP address filter
pdu = pdu.append(sheet(main,'IP','^10\..*\..*\.24[0-8]'))
main = sheet_x(main,'IP','^10\..*\..*\.24[0-8]')
```

```python
print('Separating  in CMDB and Not in CMDB data')
main_in, main_out = cmdb(main)
lb_in, lb_out = cmdb(lb)
devops_in, devops_out = cmdb(devops)
link_in, link_out = cmdb(link)
blank_in, blank_out = cmdb(blank)
dup_in, dup_out = cmdb(duplicate)
agent_in, agent_out = cmdb(agent)
ls_in, ls_out = cmdb(ls)
cy_in, cy_out = cmdb(cy)
pdu_in, pdu_out = cmdb(pdu)
pub_in, pub_out = cmdb(pub)
bo1_in, bo1_out = cmdb(bo1)
bronto_in, bronto_out = cmdb(bronto)
openair_in, openair_out = cmdb(openair)
oci_in, oci_out = cmdb(oci)
payroll_in, payroll_out = cmdb(payroll)
console_in, console_out = cmdb(console)
vpn_in, vpn_out = cmdb(vpn)
nouse_in, nouse_out = cmdb(nouse)
vm_in, vm_out = cmdb(vm)
interface_in, interface_out = cmdb(interface)
available_in, available_out = cmdb(available)
```

```python
#arrays for easy access
in_cmdb = [main_in,lb_in,devops_in,link_in, blank_in, dup_in, agent_in, ls_in, cy_in, pdu_in, pub_in, bo1_in, bronto_in, openair_in, oci_in, payroll_in, console_in, vpn_in, nouse_in, vm_in,interface_in,available_in]
in_cmdb_sheets = ['Main','Load Balancer','DevOps','Links', 'None', 'Duplicates', 'Agent', 'LS', 'CY', 'PDU', 'Public Facing', 'BO1', 'Bronto', 'OpenAir', 'OCI', 'Payroll', 'Console', 'VPN', 'NoUse', 'VM','Interface','Available']
out_cmdb = [main_out,lb_out,devops_out,link_out, blank_out, dup_out, agent_out, ls_out, cy_out, pdu_out, pub_out, bo1_out, bronto_out, openair_out, oci_out, payroll_out, console_out, vpn_out, nouse_out,vm_out,interface_out,available_out]
out_cmdb_sheets = ['Main','Load Balancer','DevOps','Links', 'None', 'Duplicates', 'Agent', 'LS', 'CY', 'PDU', 'Public Facing', 'BO1', 'Bronto', 'OpenAir', 'OCI', 'Payroll', 'Console', 'VPN', 'NoUse', 'VM','Interface','Available']
```

```python
with pd.ExcelWriter('In_CMDB_LATEST.xlsx') as writer:
    for df, sheets in zip(in_cmdb, in_cmdb_sheets):
        workbooks(df,sheets)
    print('In_CMDB_LATEST.xlsx Exported')

with pd.ExcelWriter('Out_CMDB_LATEST.xlsx') as writer:
    for df, sheets in zip(out_cmdb, out_cmdb_sheets):
        workbooks(df,sheets)
    print('Out_CMDB_LATEST.xlsx Exported')    
stop = time.perf_counter()

print("It took "+str(round((stop-start)/60,2))+" minutes for the script to complete")
```
