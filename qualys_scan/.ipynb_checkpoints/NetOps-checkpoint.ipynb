```python
import pandas as pd
import os,shutil, requests, time
import concurrent.futures
start = time.perf_counter()
```

```python
workbook = pd.ExcelFile('latest_sample_pinged_data.xlsx')
```

```python
data = pd.read_excel(workbook)
```

```python
netops = data.sort_values('Last Scanned Date',ascending=False).drop_duplicates('IP', keep='last')
```

```python
dns_filter = netops[netops['DNS'].str.contains('vlb|-qlb|lb|vip|-plb|-mip|-link|-ls|-interface|vlan|None|available', na=False)]
dns_filter_x= netops[~netops['DNS'].str.contains('vlb|-qlb|lb|vip|-plb|-mip|-link|-ls|-interface|vlan|None|available', na=False)]
```

```python
bu_filter = dns_filter_x[dns_filter_x['Tags'].str.contains('BU:NetOps',na=False)]
```

```python
newnet = dns_filter.append(bu_filter)
```

```python
not_newnet = newnet[newnet['IP in CMDB']=='IP NOT in CMDB'].sort_values('IP')
```

```python
len(not_newnet)
```

```python
with pd.ExcelWriter('NetOps_IP.xlsx') as writer:
    not_newnet.to_excel(writer,index=False)
```

```python
print("NetOps_IP has been exported")
stop = time.perf_counter()
print("It took "+str(round((stop-start)/60,2))+" minutes for the script to complete")
```
