```python
import pandas as pd
```

```python
data = pd.read_csv('management_ip_output.txt')
data.columns=['scans']
```

```python
data = data[data['scans'].str.contains('NOK dns failure')]
```

```python
# with pd.ExcelWriter('managementIPOutput.xlsx') as writer:
#     x.to_excel(writer, index=False)
```

```python
data['DNS']=''
data['CNAME']=''
data['Management_IP']=''
```

```python
def stripped(dns):
    begin, end = dns.find('('), dns.find(')')
    filtered = dns[begin: end+1]
    filtered = filtered.strip('()')
    if len(filtered) < 1:
        end = dns.find('com')
        filtered = dns[0:end+3]
        data['DNS'][data['scans'].str.contains(filtered)]=filtered
        
    else:
        data['DNS'][data['scans'].str.contains(filtered)]=filtered
        
```

```python
for x in data['scans']:
    stripped(x)
```

```python
with pd.ExcelWriter('ManagementIPOutput.xlsx') as writer:
    data.to_excel(writer, index=False)
```
