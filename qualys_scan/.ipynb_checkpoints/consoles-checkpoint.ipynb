```python
import pandas as pd
import os,shutil, requests, time
import concurrent.futures
start = time.perf_counter()
```

```python
workbook = pd.ExcelFile('latest_sample_pinged_data.xlsx')
```

```python
data = pd.read_excel(workbook)
```

```python
console = data[data['DNS'].str.contains('console', na=False)]
```

```python
console = console.sort_values('Last Scanned Date', ascending=False).drop_duplicates('IP', keep='last')
```

```python
console
```

```python
out = console[console['IP in CMDB'] == 'IP NOT in CMDB']
```

```python
out['Console DNS']=''
```

```python
def outdated(dns):
    API = '0a66e9aecdc8c49191309691d8c340b920a060b3'
    url = f'https://cmdb.netledger.com/api/dc-hosts/?hostname=Outdated%20({dns})'
    headers = {'Authorization': 'Token '+ API}
    response = requests.get(url, headers=headers)
    result = response.json()
    count = result['count']
    if count == 1:
            out['DNS in CMDB'][out['DNS'] == dns] = 'DNS in CMDB - Outdated'
            print(str(dns),'DNS in CMDB - Outdated')
    elif count < 1:
            out['DNS in CMDB'][out['DNS'] == dns] = 'DNS NOT in CMDB'
            print(str(dns)," DNS NOT in cmdb")  
    

def dns_search (dns):
    API = '0a66e9aecdc8c49191309691d8c340b920a060b3'
    url = f'https://cmdb.netledger.com/api/dc-hosts/?hostname={dns}'
    headers = {'Authorization': 'Token '+ API}
    response = requests.get(url, headers=headers)
    result = response.json()
    count = result['count']
    if count == 0:
        count = outdated(dns)
    
    elif count == 1:
        out['DNS in CMDB'][out['DNS'] == dns] = 'DNS in CMDB'
        print(str(dns)," DNS in cmdb")
    elif count > 1:
        out['DNS in CMDB'][out['DNS'] == dns] = 'DNS in CMDB + '
```

```python
def console_split(console):
    dns = console.replace('-console','')
    out['Console DNS'][out['DNS']== console] = console
    out['DNS'][out['DNS']==console]=dns
    dns_search(dns)
    
```

```python
console_split('db454-console.prod.svale.netledger.com')
```

```python
out[['IP','DNS','Console DNS','Last Scanned Date','DNS in CMDB']]
```

```python
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(console_split,out['DNS'])
```

```python
out = out[['IP','DNS','Console DNS','Last Scanned Date','DNS in CMDB']].sort_values('DNS in CMDB',ascending=False)
out
```

```python
with pd.ExcelWriter('console_dns.xlsx') as writer:
    out.to_excel(writer,index=False)
```

```python
print("console_dns has been exported")
stop = time.perf_counter()
print("It took "+str(round((stop-start)/60,2))+" minutes for the script to complete")
```
