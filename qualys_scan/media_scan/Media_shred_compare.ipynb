```python
import pandas as pd
import os
# os.chdir('media_scan')
```

```python
dc_data = pd.read_excel('dc_data.xlsx')
simms_data = pd.read_excel('simms_data.xlsx')
codd_bo1 = pd.read_excel('CODD_BO1.xlsx')
dc_data_sn = dc_data['Manufacturer Serial #']
simms_data_sn = simms_data['Device\nSerial Number']
codd_bo1_sn = codd_bo1['CODD']
```

```python
#Checking Data Counts
print('dc_data',len(dc_data_sn))
print('simms_data',len(simms_data_sn))
print('codd_data',(len(codd_bo1_sn)))
```

```python
#Checking for duplicates
if len(simms_data[simms_data.duplicated()]) > 1:
    print('Duplicate found in simms_data')
else:
    print('No duplicates found in simms_data')

    
if len(codd_bo1_sn[codd_bo1_sn.duplicated()]) > 1:
    print('Duplicate found in codd_data')
else:
    print('No duplicates found in codd_data')

if len(dc_data_sn[dc_data.duplicated()]) > 1:
    print('Duplicate found in dc_data')
    print(dc_data[dc_data.duplicated()])
else:
    print('No duplicates found in dc_data')
```

```python
# Checking SIMMS SPREADSHEET VS CODD

for x in codd_bo1_sn:
    if x not in simms_data_sn.values:
        print (x, 'not in CODD list')
```

```python
#CHECKING SIMMS SPREADSHEET VS DC SPREADSHEET
simm_not_in_dc = []
dc_dc_not_in_simm = []
for x in simms_data_sn:
    if x not in dc_data_sn.values:
        simm_not_in_dc.append(x)
print('Drives count from SIMMS not listed in DC:',len(simm_not_in_dc))
for x in dc_data_sn:
    if x not in simms_data_sn.values:
        dc_dc_not_in_simm.append(x)
print('Drives count from DC not listed in SIMMS:',len(dc_dc_not_in_simm))
```
