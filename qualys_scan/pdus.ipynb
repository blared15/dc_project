```python
import pandas as pd
import os,shutil, requests, time
import concurrent.futures
start = time.perf_counter()
```

```python
workbook = pd.ExcelFile('latest_sample_pinged_data.xlsx')
```

```python
data = pd.read_excel(workbook)
```

```python
data = data.sort_values('Last Scanned Date',ascending=False).drop_duplicates('IP', keep='last')
```

```python
pdus = data[data['DNS'].str.contains('pdu-', na=False)]
```

```python
not_pdus = pdus[pdus['DNS in CMDB']=='DNS NOT in CMDB']
```

```python
svale = not_pdus[not_pdus['DNS'].str.contains('svale',na=False)]

sea = not_pdus[not_pdus['DNS'].str.contains('sea',na=False)]
bos = not_pdus[not_pdus['DNS'].str.contains('bos',na=False)]
ams = not_pdus[not_pdus['DNS'].str.contains('ams',na=False)]
chi = not_pdus[not_pdus['DNS'].str.contains('chi',na=False)]
dub = not_pdus[not_pdus['DNS'].str.contains('dub',na=False)]
```

```python
frames = [sea,bos,ams,chi,dub]
sheets = ['SEA','BOS','AMS','CHI','DUB']
```

```python
with pd.ExcelWriter('PDUS_not_in_CMDB.xlsx') as writer:
    for frame,sheet in zip(frames,sheets):
        if len(frame) > 0:
            frame.to_excel(writer,sheet_name=sheet, index=False)
```

```python
print("PDUS_not_in_CMDB has been exported")
stop = time.perf_counter()
print("It took "+str(round((stop-start)/60,2))+" minutes for the script to complete")
```
